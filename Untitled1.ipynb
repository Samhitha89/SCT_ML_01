```python
import random

def ai_agent_response(user_input):
    user_input = user_input.lower()

    faq = {
        "hello": "Hi! I'm your Study Support Bot. How can I help you today?",
        "hi": "Hello! Ask me anything about study tips or productivity.",
        "how to study": "Try the Pomodoro technique: 25 mins study + 5 mins break.",
        "motivate me": "Small steps daily lead to big results â€” you've got this! ðŸ’ª",
        "time management": "Create a taskâ€‘priority list and work on highâ€‘value tasks first.",
        "notes tips": "Rewrite concepts in your own words â€” that improves memory retention.",
        "bye": "Goodbye! Keep learning and stay consistent âœ¨"
    }

    for key in faq:
        if key in user_input:
            return faq[key]

    return random.choice([
        "Interesting! Can you tell me a bit more?",
        "I'm still learning â€” try asking in a simpler way ðŸ˜Š",
        "Let me thinkâ€¦ maybe try rephrasing that."
    ])

def run_agent():
    print("\nðŸŽ“ Student Support AI Agent â€” Chatbot\nType 'exit' to quit.\n")
    while True:
        user = input("You: ")
        if user.lower() == "exit":
            print("Bot: Goodbye! ðŸ‘‹")
            break
        print("Bot:", ai_agent_response(user))

if __name__ == "__main__":
    run_agent()
```
